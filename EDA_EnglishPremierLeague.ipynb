# Simple Stat 3 questions

```python
print('คำถามข้อที่ 1: HomeTeam ทีมไหนชนะมากที่สุด')

#สร้างตาราง
a = df.loc[:, ['HomeTeam','FTR','HTR']]

#เพิ่ม column count FTR
a['CountFTR'] = [1 for i in range(0,len(df['HomeTeam']))] 

# Pivot เพื่อดูจำนวน ชนะ เสมอ หรือแพ้
table_a = pd.pivot_table(a, values = 'CountFTR',
                      index = ['HomeTeam'],
                      columns = ['FTR'],
                      aggfunc = 'sum',
                      margins = True) 
#margins =หาผลรวม row and column


# Sort H = Home Wins
table2_a = table_a.reindex(table_a.sort_values(by='H', ascending = False).index)

#คำตอบ
# print(table2_a)
print('---------------------------------------')
print(f"ดังนั้นในการแข่งฟุตบอลโดยที่ \n{table2_a.iloc[1:3,0:4]} \nเป็น HomeTeam และชนะไปทั้งหมด {int(table2_a.loc[('Liverpool','H')])} ครั้งต่อการแข่งทั้งหมด {table2_a.iloc[1,3]} ครั้ง")
```

```python
print('คำถามข้อที่ 2: ทีม Leeds ตอนที่เป็น HomeTeam และ AwayTeam แข่งกับทีมไหนแล้วชนะ และทำประตูได้มากที่สุด')

b = df.loc[:, ['HomeTeam','AwayTeam','FTR','HTR','FTHG','FTAG']]
# b

#Filter Leed in Pivot table
table_b = b[(b['HomeTeam'] == 'Leeds')].pivot_table(
                        values = ['FTAG','FTHG','FTR'],
                        index = ['HomeTeam','AwayTeam'],
                        aggfunc = 'sum')

#filter H ก่อน แล้ว ค่อย sort column ตามด้วย sort value
table_b_1 = table_b[table_b['FTR'] == 'H'].sort_index(axis = 1, ascending = False).sort_values(by = ['FTHG','FTAG'], ascending = False)


#Filter Leed in Pivot table
table_b2 = b[(b['AwayTeam'] == 'Leeds')].pivot_table(
                        values = ['FTAG','FTHG','FTR'],
                        index = ['AwayTeam','HomeTeam'],
                        aggfunc = 'sum')

#filter A ก่อน แล้ว ค่อย sort column ตามด้วย sort value
table_b2_1 = table_b2[table_b2['FTR'] == 'A'].sort_index(axis = 1, ascending = True).sort_values(by = ['FTHG','FTAG'], ascending = False)

print(table_b_1)
print(table_b2_1)
print('--------------------------------------------- \n')

print(f"สรุปแล้ว ทีม Leeds แข่งชนะทำประตูได้มากสุดกับทีม \n\
        \n 1. กรณีที่ทีม Leeds เป็น HomeTeam คือทีม \
        \n {table_b_1[table_b_1['FTHG'] > 2]}\
        \n \
        \n 2. กรณีที่ทีม Leeds เป็น AwayTeam คือทีม \
        \n {table_b2_1[table_b2_1['FTAG'] > 2]}")
```

```python
print(f'คำถามข้อที่ 3: ทีม Leeds ได้ใบแดงและใบเหลืองครั้งไหนมากที่สุด และผลการแข่งขันครั้งนั้นใครชนะ?')
c = df.loc[:, ['HomeTeam','AwayTeam','FTR','HTR','FTHG','FTAG','HY','AY','HR','AR']]

table_c = c[(c['HomeTeam'] == 'Leeds')].pivot_table(
                                        values = ['HY','HR','FTR'],
                                        index = ['HomeTeam', 'AwayTeam'],
                                        aggfunc = np.sum).sort_values(by=['HY','HR'],ascending = False)


# table_c2 = table_c[(table_c.index.get_level_values('AwayTeam') == 'Liverpool') | (table_c.index.get_level_values('AwayTeam') == 'Chelsea')]
# table_c2 = table_c[table_c.index.get_level_values('AwayTeam').isin(['Liverpool','Chelsea'])]
# table_c2


# print Table Home Team Yellow Max
table_c2 = table_c[table_c['HY'] == table_c['HY'].max()]

# table_c2.iloc[0,0]

print(table_c2)
print(f"เพราะฉะนั้นสรุปได้ว่าในการแข่งขันทีม Leeds ได้ใบเหลืองมากที่สุดจำนวน {table_c['HY'].max()} ครั้ง และผลการแข่งขันนั้นทีม {table_c2.iloc[0,0]} ชนะ")

# print Table Home Team Reb Max
table_c3 = table_c[table_c['HR'] == table_c['HR'].max()]

print(table_c3)
print(f"เพราะฉะนั้นสรุปได้ว่าในการแข่งขันทีม Leeds ได้ใบแดงมากที่สุดจำนวน {table_c['HR'].max()} ครั้ง และผลการแข่งขันนั้นทีม {table_c2.iloc[0,0]} ชนะ")
```
