```python
import pickle
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt #dpi 600
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/NarongkiatPW/DADS/DADS6003_Machine_Learning/Housing.csv')
df.head(10)
```

```python
df.shape
```

```python
# df.head(10)
df2 =df.copy()
# df2['mainroad'] = df2['mainroad'].apply(lambda x: '1' if x == 'yes' else '0')
df2['mainroad'] = np.where(df2['mainroad'] == 'yes', '1', '0').astype(int)
df2['guestroom'] = np.where(df2['guestroom'] == 'yes','1','0').astype(int)
df2['basement'] = np.where(df2['basement']== 'yes', '1','0').astype(int)
df2['hotwaterheating'] = np.where(df2['hotwaterheating'] =='yes','1','0').astype(int)
df2['airconditioning'] = np.where(df2['airconditioning'] == 'yes','1','0').astype(int)
df2['prefarea'] = np.where(df2['prefarea'] =='yes','1','0').astype(int)
df2['furnishingstatus'] = np.where(df2['furnishingstatus'] =='furnished','1',np.where(df2['furnishingstatus'] =='semi-furnished','0.5','0')).astype(float)
df2.head(20)
# print(df.shape)
```

```python
# กำหนด x และ y
col_x = ['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']
x = np.array(df2[col_x])
y = np.array(df2['price'])

# Function approximation f:x->y, f:R(n rows*d column)-> R (n rows*1 column)
# x_b= np.c_[np.ones((x.shape[0],1)),x] # n rows*1 col โดยใช้ np.ones บอก row*col และ np.c_ -> concat 1:1  ของตัวหน้ากับตัวหลัง

# หา theta โดยการ call inverse function from linear algebra module
# theta = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
# theta
```

# Multiple Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Split data train 80%, test 20%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# สร้าง Model
lin_reg = LinearRegression()
# Train Model
lin_reg.fit(x_train,y_train)
print(lin_reg.intercept_, " , ",lin_reg.coef_)

#Prediction
y_predict = lin_reg.predict(x_test)
print(f"x_test.shape = {x_test.shape}")
print(f"y_predict = {y_predict}")
```

```python
# หา MSE
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(y_test, y_predict)
print(f"MSE = {mean_squared_error(y_test, y_predict)}")

# print(f"y.shape = {y.shape}")
# print(f"y_predict.shape = {y_predict.shape}")
# print(f"y_train.shape = {y_train.shape}")
# print(f"y_test.shape = {y_test.shape}")
```
